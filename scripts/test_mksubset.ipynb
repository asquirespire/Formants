```python
import pandas as pd
from sklearn.utils import shuffle
import os
```

```python
FORMANTS_CSV_PATH = "/home/jeevan/Jeevan_K/Projects/Asquire/Reverb-Quest/Formants/scripts/all_timit_vowel_formants_mf_p.csv"
klatt_json_data_path = "/home/jeevan/Jeevan_K/Projects/Asquire/Reverb-Quest/Formants/CSV/for_klatt"
```

```python
all_phone_formants_df = pd.read_csv(FORMANTS_CSV_PATH); all_phone_formants_df
```

```python
all_vowels = pd.unique(all_phone_formants_df["phone"]); all_vowels
```

```python
LIMIT = 200
dfs = []
for vowel in all_vowels:
    sub_df = all_phone_formants_df[all_phone_formants_df["phone"] == vowel]

    sub_df_male = shuffle(sub_df[sub_df["sex"] == "M"])
    sub_df_female =  shuffle(sub_df[sub_df["sex"] == "F"])

    size_m = LIMIT if len(sub_df_male) > LIMIT else len(sub_df_male)
    size_f = LIMIT if len(sub_df_female) > LIMIT else len(sub_df_female)

    sub_df_male = sub_df_male[:size_m]
    sub_df_female = sub_df_female[:size_f]
    
    df = pd.concat([sub_df_male, sub_df_female])

    dfs.append(df)

    
timit_vowel_subset = pd.concat(dfs)
timit_vowel_subset = timit_vowel_subset.reset_index()
timit_vowel_subset = timit_vowel_subset.drop(columns=["id"])
timit_vowel_subset.index.name = "idx"
timit_vowel_subset = timit_vowel_subset.drop(columns=["index"])
pth_json = os.path.join(klatt_json_data_path, "timit_vowel_subset_mf_p.json")
timit_vowel_subset.to_json(pth_json, index=True, orient="table")
timit_vowel_subset.to_csv("timit_vowel_subset_mf_p.csv", index=True)

timit_vowel_subset
```
